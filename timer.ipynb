# Cora

```python
data = json.load(open('./timer_cora.json', 'r'))
```

```python
data['train']['epoch 0']
```

```python
epochs = range(50)
times = list(data["train"].values())

average_time = sum(times) / len(times)

plt.plot(epochs, times, marker='o', label='Epoch Time')
plt.axhline(y=average_time, color='r', linestyle='--', label=f'Average Time: {average_time:.2f}')
plt.xlabel('Epoch')
plt.ylabel('Time (seconds)')
plt.title('Training time (cora)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
